```python
Qual bairro + culinária tem a melhor nota?
melhor_por_bairro = cuisine_rating.groupby(["Location","Cuisines"])["Overall Rating"].mean().round(2)
melhor_por_bairro = melhor_por_bairro.sort_values(ascending=False)

print("TOP 10 MELHORES COMBINAÇÕES:")
print(melhor_por_bairro.head(10))
```

```python
# Gráfico 1: Heatmap das melhores culinárias por bairro
pivot = cuisine_rating.pivot_table(values='Overall Rating',
                       index='Location',
                       columns='Cuisines',
                       aggfunc='mean')

plt.figure(figsize=(12, 8))
sns.heatmap(pivot, annot=True, cmap="YlGnBu", fmt=".2f", linewidths=.5)
plt.title('Média de Avaliação Geral por Bairro e Tipo de Culinária', fontsize=16, pad=20)
plt.ylabel('Bairro')
plt.xlabel('Culinária')
plt.tight_layout()
plt.show()
```

```python
# Gráfico 2: Top 10 melhores lugares
top10 = melhor_por_bairro.head(10)

plt.figure(figsize=(10,6))
top10.plot(kind="barh", color="lightblue")
plt.title("Top 10 Melhores Avaliações - Bairro + Culinária")
plt.xlabel("Nota Média")
plt.ylabel("")
plt.gca().invert_yaxis()
plt.show()
```

```python
# Gráfico 3: Top 10 melhores restaurantes (culinária + bairro)
top_by_location = cuisine_rating.groupby(['Location', 'Cuisines'])['Overall Rating'].mean().sort_values(ascending=False)
top10 = top_by_location.head(10).round(2)

plt.figure(figsize=(10, 6))
ax = top10.plot(kind='barh', color='skyblue', edgecolor='navy', alpha=0.8)
plt.title('Top 10 Combinações Bairro + Culinária com Melhores Avaliações', fontsize=14, pad=20)
plt.xlabel('Avaliação Média Geral')
plt.ylabel('')
for i, v in enumerate(top10.values):
    plt.text(v + 0.05, i, str(v), va='center')
plt.tight_layout()
plt.show()
```

```python
# Gráfico 4: Distribuição de avaliações por culinária
plt.figure(figsize=(12, 5))
sns.boxplot(data=cuisine_rating, x='Cuisines', y='Overall Rating', order=cuisine_rating.groupby('Cuisines')['Overall Rating'].median().sort_values(ascending=False).index)
plt.title('Distribuição das Avaliações por Tipo de Culinária', fontsize=14, pad=20)
plt.xticks(rotation=45)
plt.ylabel('Avaliação Geral')
plt.xlabel('Culinária')
plt.tight_layout()
plt.show()
```

```python
# Gráfico 5: Avaliação média por faixa etária e culinária (insight de negócio forte!)

# Calculate age from YOB (assuming current year is 2024)
cuisine_rating['age'] = 2024 - cuisine_rating['YOB']

cuisine_rating['age_group'] = pd.cut(cuisine_rating['age'], bins=[0, 25, 35, 50, 100], labels=['<25', '25-35', '35-50', '50+'])

plt.figure(figsize=(12, 6))
sns.barplot(data=cuisine_rating, x='Cuisines', y='Overall Rating', hue='age_group', ci=None)
plt.title('Preferência de Culinária por Faixa Etária', fontsize=14, pad=20)
plt.xlabel('Culinária')
plt.ylabel('Avaliação Média')
plt.legend(title='Idade')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Gráfico 6: Correlação entre variáveis numéricas
numerical = cuisine_rating[['Budget', 'Food Rating', 'Service Rating', 'Overall Rating', 'age']]
corr = numerical.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', center=0, square=True)
plt.title('Correlação entre Variáveis Numéricas')
plt.show()
```
